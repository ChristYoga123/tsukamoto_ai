# FUZZIFIKASI

```python
# Mobil
def mobilBanyak(mobil):
    if mobil >= 50:
        return 1
    elif mobil <= 50 and mobil >= 49:
        return ((50 - mobil) / (50 - 49))
    else:
        return 0

def mobilSedikit(mobil):
    if mobil <= 30:
        return 1
    elif mobil <= 30 and mobil >= 29:
        return ((30 - mobil) / (30 - 29))
    else:
        return 0

def mobilSedang(mobil):
    if mobil >= 14 and mobil <= 15:
        return ((15 - mobil) / (15 - 14))
    elif mobil >= 15 and mobil <= 65:
        return 1
    elif mobil >= 64 and mobil <= 65:
        return ((mobil - 64) / (65 - 64))
    else:
        return 0

# Motor
def motorBanyak(motor):
    if motor >= 50:
        return 1
    elif motor <= 50 and motor >= 49:
        return ((50 - motor) / (50 - 49))
    else:
        return 0

def motorSedikit(motor):
    if motor <= 40:
        return 1
    elif motor <= 40 and motor >= 39:
        return ((40 - motor) / (40 - 39))
    else:
        return 0

def motorSedang(motor):
    if motor >= 19 and motor <= 20:
        return ((20 - motor) / (20 - 19))
    elif motor >= 20 and motor <= 70:
        return 1
    elif motor >= 69 and motor <= 70:
        return ((motor - 69) / (70 - 69))
    else:
        return 0

# Lebar Jalan
def jalanLebar(jalan):
    if jalan > 8:
        return 1
    elif jalan < 8:
        return 0
    else:
        return ((jalan - 8) / (9 - 8))

def jalanSempit(jalan):
    if jalan < 8:
        return 1
    elif jalan > 8:
        return 0
    else:
        ((8 - jalan) / (8 - 7))

def jalanSedang(jalan):
    if jalan <= 6 and jalan >= 5:
        return ((6 - jalan) / (6 - 5))
    elif jalan >= 6 and jalan <= 10:
        return 1
    elif jalan >= 9 and jalan <= 10:
        return ((10 - jalan) / (10 - 9))
```
